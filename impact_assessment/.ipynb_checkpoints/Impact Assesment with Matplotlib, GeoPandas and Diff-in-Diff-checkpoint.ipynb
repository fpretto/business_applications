#### Spatial Autocorrelation

##### Spatial Similarity

```python
wq =  lp.weights.Queen.from_dataframe(gdf_barrios)
wq.transform = 'r'
```

##### Attribute Similarity

```python
y = gdf_barrios['precio_alq_prom']
ylag = lp.weights.lag_spatial(wq, y)
ylagq5 = mc.Quantiles(ylag, k=5)
```

```python
f,ax = plt.subplots(1, figsize=(9, 9), subplot_kw=dict(aspect='equal'))
gdf_barrios.assign(cl=ylagq5.yb).plot(column='cl', categorical=True,
        k=5, cmap='GnBu', linewidth=0.1,
        edgecolor='white', legend=True, ax=ax)
ax.set_axis_off()
plt.title("Spatial Lag Median Price (Quantiles)")

plt.show()
```

```python
gdf_barrios['lag_precio_alq_prom'] = ylag
f,ax = plt.subplots(1,2,figsize=(2.16*4,4), subplot_kw=dict(aspect='equal'))
gdf_barrios.plot(column='precio_alq_prom', ax=ax[0], edgecolor='k', scheme="quantiles",  k=5, cmap='GnBu')
ax[0].axis(gdf_barrios.total_bounds[np.asarray([0,2,1,3])])
ax[0].set_title("Price")
gdf_barrios.plot(column='lag_precio_alq_prom', ax=ax[1], edgecolor='k', scheme='quantiles', cmap='GnBu', k=5)
ax[1].axis(gdf_barrios.total_bounds[np.asarray([0,2,1,3])])
ax[1].set_title("Spatial Lag Price")
ax[0].axis('off')
ax[1].axis('off')
plt.show()
```

### Evaluacion de Impacto

Para realizar este punto se decidio trabajar con datos del **Sistema Unico de Atencion Ciudadana (SUACI)** de la ciudad de Buenos Aires. Este dataset contiene las llamadas de los ciudadanos al SUACI agrupados por Periodo, Tipo de prestacion, Comuna y Barrio.

Ademas, se utilizo el dataset de que contiene el poligono de cada comuna de GCBA.

```python
# Cargamos el dataset de las comunas
gdf_barrios = gpd.read_file(path+'barrios.csv')

# Cargamos el dataset de SUACI
df_suaci = pd.read_csv(path+'suaci_comunas.csv', sep=',', decimal='.', encoding='latin-1')
df_suaci = df_suaci[~df_suaci['COMUNA'].isna()]
df_suaci['COMUNA'] = df_suaci['COMUNA'].apply(lambda x: str(int(x)))
df_suaci['PERIODO'] = df_suaci['PERIODO'].apply(lambda x: str(x))

# Mergeamos ambos datasets y transformamos el DataFrame de SUACI a GeoDataFrame
df_suaci = df_suaci.merge(gdf_barrios[['barrio','comuna', 'geometry']], right_on=['barrio','comuna'], left_on=['BARRIO','COMUNA'], how='left')
gdf_suaci = gpd.GeoDataFrame(df_suaci, geometry='geometry')

gdf_suaci['Mes'] = gdf_suaci['PERIODO'].apply(lambda x: int(x[4:6]))
gdf_suaci['Anio'] = gdf_suaci['PERIODO'].apply(lambda x: int(x[0:4]))
```

#### Definicion de solucion y funciones auxiliares

Se buscará crear una herramienta gráfica sobra la cual poder realizar análisis de evaluación de impacto sobre mejoras en la atención a los ciudadanos.

Para esto, se definen 2 funciones:

1- **plot_map**: grafica el total de prestaciones realizadas por comuna y periodo en el mapa de GCBA. Se puede filtrar el tipo de prestacion, año y mes.

2- **plot_line**: gráfico de línea con la evolución de la cantidad de prestaciones realizadas. Se puede filtrar por tipo de prestacion y barrio.

```python
def plot_map(Mes=1, Anio=2013, Prestacion='TODAS'):
    
    # Mapa
    gdf_plot=gdf_suaci[(gdf_suaci['Mes']==Mes) & (gdf_suaci['Anio']==Anio)]
    
    if Prestacion != 'TODAS':
        gdf_plot=gdf_plot[gdf_plot['TIPO_PRESTACION'] == Prestacion].copy()
        
    # Plot
    fig, ax = plt.subplots(figsize=(24,10), subplot_kw=dict(aspect='equal'))
    gdf_plot.plot(column='total', scheme='Quantiles', k=15, cmap='GnBu', ax=ax, legend=True)
```

```python
def plot_line(Prestacion='TODAS', BarrioTrat='TODOS', BarrioCtrl='TODOS'):
    
    # Grafico de Linea 1 (Barrio_Tratamiento)
    if (Prestacion == 'TODAS') & (BarrioTrat == 'TODOS'):
        gdf_plot1 = gdf_suaci[['PERIODO', 'total']].groupby('PERIODO').sum()
    
    elif (Prestacion != 'TODAS') & (BarrioTrat != 'TODOS'):
        gdf_filtered = gdf_suaci[(gdf_suaci['TIPO_PRESTACION'] == Prestacion) & (gdf_suaci['BARRIO'] == BarrioTrat)].copy()
        gdf_plot1 = gdf_filtered[['PERIODO', 'total']].groupby('PERIODO').sum()
        
    elif (Prestacion != 'TODAS') & (BarrioTrat == 'TODOS'):
        gdf_filtered = gdf_suaci[gdf_suaci['TIPO_PRESTACION'] == Prestacion].copy()
        gdf_plot1 = gdf_filtered[['PERIODO', 'total']].groupby('PERIODO').sum()
        
    elif (Prestacion == 'TODAS') & (BarrioTrat != 'TODOS'):
        gdf_filtered = gdf_suaci[gdf_suaci['BARRIO'] == BarrioTrat].copy()
        gdf_plot1 = gdf_filtered[['PERIODO', 'total']].groupby('PERIODO').sum()
        
    
    # Grafico de Linea 2 (Barrio_Control)
    if (Prestacion == 'TODAS') & (BarrioCtrl == 'TODOS'):
        gdf_plot2 = gdf_suaci[['PERIODO', 'total']].groupby('PERIODO').sum()
    
    elif (Prestacion != 'TODAS') & (BarrioCtrl != 'TODOS'):
        gdf_filtered = gdf_suaci[(gdf_suaci['TIPO_PRESTACION'] == Prestacion) & (gdf_suaci['BARRIO'] == BarrioCtrl)].copy()
        gdf_plot2 = gdf_filtered[['PERIODO', 'total']].groupby('PERIODO').sum()
        
    elif (Prestacion != 'TODAS') & (BarrioCtrl == 'TODOS'):
        gdf_filtered = gdf_suaci[gdf_suaci['TIPO_PRESTACION'] == Prestacion].copy()
        gdf_plot2 = gdf_filtered[['PERIODO', 'total']].groupby('PERIODO').sum()
        
    elif (Prestacion == 'TODAS') & (BarrioCtrl != 'TODOS'):
        gdf_filtered = gdf_suaci[gdf_suaci['BARRIO'] == BarrioCtrl].copy()
        gdf_plot2 = gdf_filtered[['PERIODO', 'total']].groupby('PERIODO').sum()
            
        
    # Plot
    fig, ax = plt.subplots(figsize=(12,6))
    gdf_plot1.plot(color='steelblue', ax=ax)
    gdf_plot2.plot(color='tomato', ax=ax)
    ax.set_ylabel('Cantidad de Prestaciones')
    ax.grid(linestyle='--')
    ax.legend(('Tratamiento: ' + BarrioTrat, 'Control: ' + BarrioCtrl))
    if (Prestacion == 'TODAS') & (BarrioTrat == 'TODOS') & (BarrioCtrl == 'TODOS'):
        plt.title('Evolucion de Prestaciones en GCBA')
    elif (Prestacion != 'TODAS') & ((BarrioTrat != 'TODOS') | (BarrioCtrl != 'TODOS')):
        plt.title('Evolucion de {} en {} y {}'.format(Prestacion, BarrioTrat, BarrioCtrl))
    elif (Prestacion != 'TODAS') & (BarrioTrat == 'TODOS') & (BarrioCtrl == 'TODOS'):
        plt.title('Evolucion de {} en GCBA'.format(Prestacion))
    elif (Prestacion == 'TODAS') & ((BarrioTrat != 'TODOS') | (BarrioTrat != 'TODOS')):
        plt.title('Evolucion de Prestaciones en {} y {}'.format(BarrioTrat, BarrioCtrl))
```

```python
## Definicion de categorias
Prestaciones=['TODAS', 'RECLAMO', 'SOLICITUD', 'DENUNCIA', 'QUEJA', 'TRAMITE']

Barrios=['TODOS','AGRONOMIA', 'ALMAGRO', 'BALVANERA', 'BARRACAS', 'BELGRANO',
         'BOCA', 'BOEDO', 'CABALLITO', 'CHACARITA', 'COGHLAN', 'COLEGIALES',
         'CONSTITUCION', 'FLORES', 'FLORESTA', 'LINIERS', 'MATADEROS',
         'MONSERRAT', 'MONTE CASTRO', 'NUÃ\x91EZ', 'NUEVA POMPEYA',
         'PALERMO', 'PARQUE AVELLANEDA', 'PARQUE CHACABUCO', 'PARQUE CHAS',
         'PARQUE PATRICIOS', 'PATERNAL', 'PUERTO MADERO', 'RECOLETA',
         'RETIRO', 'SAAVEDRA', 'SAN CRISTOBAL', 'SAN NICOLAS', 'SAN TELMO',
         'VELEZ SARSFIELD', 'VERSALLES', 'VILLA CRESPO', 'VILLA DEL PARQUE',
         'VILLA DEVOTO', 'VILLA GRAL. MITRE', 'VILLA LUGANO', 'VILLA LURO',
         'VILLA ORTUZAR', 'VILLA PUEYRREDON', 'VILLA REAL',
         'VILLA RIACHUELO', 'VILLA SANTA RITA', 'VILLA SOLDATI', 'VILLA URQUIZA']
```

#### Herramienta de Analisis

Creemos que la herramienta desarrollada es una buena solución para realizar análisis de evaluación de impacto. La misma consta de 2 widgets interactivos:

**Widget Mapa**

En el primer widget, se puede analizar interactivamente en el mapa la evolución de cada tipo de prestación en toda la ciudad de Buenos Aires, barrio por barrio. Esto permite hacer comparaciones entre barrios o detectar zonas "calientes" donde se concentra algún tipo de prestación. Por ejemplo, se puede detectar cuales son los barrios de donde provienen la mayor proporción de RECLAMOS o DENUNCIAS, de manera de diseñar políticas públicas focalizadas en dichos barrios.

**Widget Linea**

Una vez detectada el área de mejora e implementada la nueva política, se puede utilizar el segundo widget para evaluar su impacto. Para esto, debe definirse primero un Grupo Control que sirva como comparativo de lo que hubiese ocurrido si no se aplicaba la política y poder así cuantificar la mejora obtenida. Este Grupo Control puede ser otro barrio de comportamiento similar y comparable al barrio de "Tratamiento". La determinación del barrio apropiado como Grupo Control se puede realizar con técnicas como *Propensity Score Matching*. Una vez matcheados los grupos Tratamiento y Control, teniendo ambos gráficos de línea se puede utilizar la técnica de *Difference-in-Difference* para evaluar visualmente el impacto de la política pública implementada. Para cuantificar la magnitud del impacto, se deberá fittear una regresión que tenga en cuenta la variable temporal, el grupo (Tratamiento/Control) y la interacción entre ambas, así como también otras variables de control:

$$y = β0 + β1 * Time + β2 * GroupID + β3 * Time : GroupID + βn * Controln + ε$$

El coeficiente β3, si resulta estadísticamente significativo, indicará la magnitud del efecto generado a partir de la implementación de la nueva política pública.

```python
# Widget Mapa
interact(plot_map, Mes=(1, 12, 1), Anio=[2013, 2014, 2015], Prestacion=Prestaciones)

# Widget Grafico de Linea
interact(plot_line, Prestacion=Prestaciones, BarrioTrat=Barrios, BarrioCtrl=Barrios)
```
